```python
#khai báo các thư viện cần thiết cho việc cào dữ liệu
import numpy as np
from selenium import webdriver
from time import sleep
import random
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.common.by import By
import pandas as pd
import requests as rq
```

```python
# tạo 1 biến url để lưu link web
url = "https://cafef.vn/thi-truong-chung-khoan.chn"
```

```python
# Gửi yêu cầu GET đến trang web và lấy nội dung
content = rq.get(url).content
print(content)
```

```python
# Lưu dữ liệu vào tệp tin
with open("data.html", "wb") as file:
    file.write(content)
```

```python
# khởi tạo một phiên bản trình duyệt Chrome
driver = webdriver.Chrome()
```

```python
# truy cập vào trang web chứa link trong chrome
driver.get(url)
```

```python
# Tìm kiếm các thẻ div có class là tlitem-flex với thẻ href
# để lấy ra link + tiêu đề
elems = driver.find_elements(By.CSS_SELECTOR,".tlitem-flex [href]")
```

```python
# in thử để kiểm tra
elems[0]
```

```python
# tạo ra 1 mảng để lưu tiêu đề các bài chứng khoáng
title = []
for i in elems:
    title.append(i.get_attribute('title'))    
```

```python
# in thử để kiểm tra
title[0]
```

```python
# tạo ra 1 mảng để lưu link các bài chứng khoáng
links = []
for i in elems:
    links.append(i.get_attribute('href'))   
```

```python
# in thử để kiểm tra
links[0]
```

```python
# Tìm kiếm các thẻ div có class là knswli-right với thẻ title
# để lấy ra thời gian
get_time = driver.find_elements(By.CSS_SELECTOR,".knswli-right [title]")
```

```python
# in thử để kiểm tra
get_time[0]
```

```python
# tạo ra 1 mảng để lưu thời gian các bài chứng khoáng
times = []
for i in get_time:
    times.append(i.get_attribute('title'))
```

```python
# in thử để kiểm tra
times[0]
```

```python
# Tìm kiếm các thẻ div có class là knswli-right với thẻ data-trimline
# để lấy ra nội dung tóm tắt
get_nd = driver.find_elements(By.CSS_SELECTOR,".knswli-right [data-trimline]")
```

```python
# tạo ra 1 mảng để lưu nội dung tóm tắt các bài chứng khoáng
nd_tomtat = []
for i in get_nd:
    nd_tomtat.append(i.text)
```

```python
# in thử để kiểm tra
nd_tomtat[0]
```

```python
# tạo 1 dataFrame từ các mảng trên
data = {'time':times, 'title':title, 'noi dung':nd_tomtat, 'link':links}
df = pd.DataFrame(data)
df
```

# Để lấy ra tất cả các nội dung từ 1 bài báo
# Ý tưởng:
- tạo 1 vòng lặp lần lượt truy cập vào link các bài báo
- truy cập vào các bài báo thông qua driver.get("link các bài báo") để lấy dữ liệu cần
- cuối cùng tạo 1 mảng để lưu các nội dung đó

```python
# tạo ra một hàm để lấy nội dung từ các link
def get_links(link):
    driver.get(link)
    sleep(random.randint(10,15)) # cần thời gian chờ để lấy nội dung trước khi next page
    x = 1
    nd_list = [] # tạo ra danh sách để lưu các nội dung
    while True:
        try: 
            nd1 = driver.find_element("xpath", "/html/body/div[3]/div[2]/div[1]/div/div[1]/div[5]/div[5]/div/div/p[{}]".format(x))
            # trỏ đến nội dung cần lấy qua xpath
            nd_list.append(nd1.text)
            x = x+1
        except NoSuchElementException:
            print('het page')
            break
    return nd_list
# vòng while sẽ chạy từ đầu đến cuối page đến khi bắt được ngoại lệ thì dừng
```

```python
List_chitiet = [] # tạo ra danh sach noi dung chi tiet
```

```python
# lần lượt truy cập vào các link để lấy nội dung
for link in links:
    print(link)
    List_chitiet.append(get_links(link))
```

```python
# in thử để kiểm tra
List_chitiet[0]
```

```python
# thêm cột List_chitiet vào bảng
df['nd_chitiet'] = List_chitiet
```
